```python
#This could be shortened but I used the long method to seperate everything 
name = input("Please enter your full name:")
CourseName = input("Please enter your course:")
Pg = input("Please enter your Prelim Grade:")
Mg = input("Please enter your Midterm Grade:")
Fg = input("Please enter your Final Grade:")

## Formula
print("----------------------------------------")
PrelimG = float (Pg)*30/100
MidG    = float (Mg)*30/100
FinalG  = float (Fg)*40/100

SemestralGrade = (PrelimG + MidG + FinalG)

#Print
print(name)
print(CourseName)
print(PrelimG)
print(MidG)
print(FinalG)
print("Your semestral grade is :" , SemestralGrade)
```
