###Create a NumPy program that performs the operations of the given matrices:

sum = A +B
difference1 = B - A
difference2 = A - B

$$ A = \begin{bmatrix}-5&0\\4&1\end{bmatrix}, B = \begin{bmatrix}6&-3\\2&3\end{bmatrix} $$

```python
import numpy as np
print("Matrix A:")
A = np.array([
     [-5,0],
     [4,1]         
])

B = np.array([
     [6,-3],
     [2,3]
])

print(A)
print("Matrix B:")
print(B)

#For SUM
sum = A+B
print("The sum of Matrix A and B:")
print(sum)
#For Difference1
diff1 = B-A
print("The difference1 of Matrix B and A:")
print(diff1)
#For Difference2
diff2 = A-B
print("The difference2 of Matrix A and B:")
print(diff2)
```
