#Determinants of Matrix

##Determinants of Matrix using numpy.linalg.det

```python
import numpy as np
print("For Array A")
A = np.array([
    [1,2,-1],
    [4,6,-2],
    [-1,3,3]
])
print(A)
print("LinAlg For A")
print(round(np.linalg.det(A)))
```
